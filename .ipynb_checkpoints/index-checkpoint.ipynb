```python
#Superponemos las provincias, ríos, ciudades, lagos y carreteras
base = provinces.plot(facecolor="azure", edgecolor='black', linewidth=0.1,figsize=(12,12))
cities.plot(marker='.', color='mediumvioletred', markersize=45,alpha=1,
            ax=base)
rivers.plot(edgecolor='darkblue', linewidth=0.4,
            ax=base)
lakes.plot(facecolor="cornsilk",edgecolor='gray', linewidth=0.4,
            ax=base)
road.plot(edgecolor='green', linewidth=0.4,
            ax=base)
```

```python
import folium

# Crear un objeto de mapa
m = folium.Map(zoom_start=10)

# Agregar las capas al mapa
folium.GeoJson(cities.to_json(), name='Cities').add_to(m)
folium.GeoJson(rivers.to_json(), name='Rivers').add_to(m)
folium.GeoJson(lakes.to_json(), name='Lakes',
               style_function=lambda x: {'fillColor': 'green', 'color': 'green'}).add_to(m)
folium.GeoJson(road.to_json(), name='Roads', style_function=lambda feature: {'color': 'red'}).add_to(m)

# Mostrar el mapa
m
```

```python
#Seleccionamos una provincia
quebec=provinces[provinces.PROVINCIA=='Quebec']
```

```python
quebec
```

```python
#Mostramos el mapa de una provincia
quebec.plot()
```

```python
#Para quedarnos solo con la parte del mapa de la provincia seleccionada 
cities_clipped = cities.clip(quebec)
rivers_clipped = rivers.clip(quebec)
lakes_clipped = lakes.clip(quebec)
road_clipped = road.clip(quebec)
```

```python
#Ajustamos los colores, líneas, puntos y tipo de línea
base = quebec.plot(facecolor="lightcyan", edgecolor='gray', linewidth=0.4,figsize=(5,5))
cities_clipped.plot(marker='*', color='deeppink', markersize=15,ax=base)
rivers_clipped.plot(edgecolor='steelblue', linewidth=0.3,ax=base)
lakes_clipped.plot(facecolor="thistle", edgecolor='gray', linewidth=0.4,figsize=(5,5), markersize=15,ax=base)
road_clipped.plot(edgecolor='mediumpurple', linewidth=0.5,ax=base)
```

```python
#Buscamos manualmente las coordenadas de la provincia que escogimos
quebecCoord=[46.81228, -71.21454]
```

```python
#Lo mostramos en el mapa interactivo
m = folium.Map(location=quebecCoord, zoom_start=4.5, tiles='CartoDB positron')

folium.GeoJson(cities_clipped.__geo_interface__, name='Cities', style_function=lambda x: {'color': 'red'}).add_to(m)
folium.GeoJson(rivers_clipped.__geo_interface__, name='Rivers', style_function=lambda x: {'color': 'blue'}).add_to(m)
folium.GeoJson(lakes_clipped.__geo_interface__, name='Lakes', style_function=lambda x: {'color': 'skyblue'}).add_to(m)
folium.GeoJson(road_clipped.__geo_interface__, name='Road', style_function=lambda x: {'color': 'green'}).add_to(m)

m
```

```python
#guardaremos el mapa como una imagen
import matplotlib.pyplot as plt

base = quebec.plot(facecolor="lightcyan", edgecolor='gray', linewidth=0.4,figsize=(5,5))
map2=cities_clipped.plot(marker='*', color='deeppink', markersize=15,ax=base)
map3=rivers_clipped.plot(edgecolor='steelblue', linewidth=0.3,ax=base)
map4=lakes_clipped.plot(facecolor="thistle", edgecolor='gray', linewidth=0.4,figsize=(5,5), markersize=15,ax=base)
mapEnd=road_clipped.plot(edgecolor='mediumpurple', linewidth=0.5,ax=base)
plt.savefig(os.path.join("maps",'mapEnd.jpg'))
```

```python
#Luego de subirlo a Github generamos el link que nos mande a dicho archivo
countryMaps='https://github.com/AnaLuciaChamorro/geoCanadadf/raw/main/maps/countryMaps.gpkg'
```

```python
from  fiona import listlayers

listlayers(countryMaps)
```

```python
provinces=gpd.read_file(countryMaps,layer='provinces')
cities=gpd.read_file(countryMaps,layer='cities')
rivers=gpd.read_file(countryMaps,layer='rivers')
lakes=gpd.read_file(countryMaps,layer='lakes')
road=gpd.read_file(countryMaps,layer='road')
```

```python
base = provinces.plot(facecolor='gainsboro')
cities.plot(ax=base, markersize=5, color='hotpink') 
rivers.plot(ax=base, linewidth=0.5)
lakes.plot(ax=base, facecolor='skyblue') 
road.plot(ax=base, linewidth=0.5,color='green',linestyle='dashed')
```
